**2.Which sport is having most number of Gold Medals so far?(Top 5)**

```python
Sport_by_medal = df[(df["Medal"] == "Gold")][["Sport", "Medal"]]
Sport_by_medal = Sport_by_medal.value_counts(["Sport"]).reset_index(name = "Count of Gold Medals")[:5]
Sport_by_medal
```

```python
sns.set_style('whitegrid')
matplotlib.rcParams['figure.figsize'] = (10,8)

plt.title("Sport with Maximum Gold Medals")
barplot = sns.barplot(data=Sport_by_medal, x="Sport", y="Count of Gold Medals",hue="Sport",palette="rainbow", dodge=False)
barplot.set_xticklabels(rotation=50, labels=Sport_by_medal.Sport)
plt.ylabel("Count of Gold Medals")
plt.show()
```

*Therefore, we can see that top 5 sports with maximum gold medals are Aquatics(1421 gold medals), Athletics(1215 gold medals), Rowing(890 gold medals), Gymnastics(820 gold medals) and Fencing(552 gold medals)*

**7. Which event is most popular in terms on number of players? (Top 5)**

**8. Which sport is having most female Gold Medalists? (Top 5)**

```python
Female_medal = df [(df ["Medal"] == "Gold") & (df ["Gender"] == "Women")][["Sport", "Medal"]]
Female_medal = Female_medal.value_counts(["Sport"]).reset_index(name = "Count of Gold Medals")[:5]
Female_medal
```

```python
sns.set_style('whitegrid')
matplotlib.rcParams['figure.figsize'] = (10,8)

plt.title("Sport With Maximum Number Of Female Gold Medalists")
barplot = sns.barplot(data=Female_medal, x="Sport", y="Count of Gold Medals", hue="Sport", palette="tab10", dodge=False)
barplot.set_xticklabels(rotation=50, labels=Female_medal.Sport)
plt.ylabel("Count of Female Gold Medalists")
plt.show()
```

*Therefore, we understand that the top five sports with maximum female gold medalists are Aquatics(589 gold medals), Athletics(389 gold medals), Gymnastics(268 gold medals), Rowing(217 gold medals) and Volleyball(166 gold medals)*
